```python
df = pd.DataFrame(np.random.randint(0,101,20).reshape(4,5),
                  columns=list('ABCDE'), index=list('wxyz'))
df
```

```python
col1 = df.columns[4].to_list()
col2 = df.columns[:3].to_list()
col3 = df.columns[5:].to_list()
new_col = col1 + col2 + col3
df = df[new_col]
```

```python
def change_col(df, col1, col2):
    return df[[col1 if c==col2 else col2 if c==col1 else c for c in df.columns]]
print(change_col(df, 'Country Code'))
```
