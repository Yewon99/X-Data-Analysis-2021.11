b. 이상치를 제거하고 위의 4가지 항목에 대해서 평균, 표준편차를 구하시오.

```python
s = iris[iris.species == 'setosa']['sepal_width']
Q1 = s.quantile(0.25)
Q3 = s.quantile(0.75)
IQR = Q3 - Q1
Q1, Q3, IQR
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
s[(s < lower)|(s > upper)] = np.nan     # '|' 는 or의 의미
s.isna().sum()
```

```python
s.mean(), s.std()
```

```python
def get_new_stat(s):
        Q1 = np.quantile(s, 0.25)
        Q3 = np.quantile(s, 0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        s[(s < lower) | (s > upper)] = np.nan
        return s.isna().sum(), s.mean(), s.std()
```

```python
iris.species.unique()
```

```python
iris.columns
```

```python
for species in iris.species.unique():
    for feature in iris.columns[:-1]:
        s = iris[iris.species == species][feature]
        outlier, new_mean, new_std = get_new_stat(s)
        print(outlier, new_mean, new_std)
```

```python
all = []
for species in iris.species.unique():
    for feature in iris.columns[:-1]:
        s = iris[iris.species == species][feature]
        old_mean, old_std = round(s.mean(),3), round(s.std(),4)
        outlier, new_mean, new_std = get_new_stat(s)
        d = {'species':species, 'feature':feature, 'old mean':old_mean, 'old std':old_std, 'outlier갯수':outlier, 'new mean':round(new_mean,3), 'new std':round(new_std,4)}
        all.append(d)
df = pd.DataFrame(all)
df.set_index(['species','feature'], inplace=True)
df
```

### 2. Titanic

a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. category1 카테고리는 다음과 같이 정의됨

- 20살이 넘으면 성별을 그대로 사용한다.
- 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
titanic.tail()
```

b. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고치시오.

```python
titanic['age'] = titanic.age.fillna(titanic.age.mean())
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)
titanic.tail()
```

### 3. MPG

a. 배기량(displacement) 대비 마력(horsepower) 열(hp_per_cc)을 추가하시오.

```python
mpg['hp_per_cc'] = mpg.horsepower / mpg.displacement
mpg.head()
```

b. name으로부터 manufacturer(제조사)와 모델을 추출하여 새로운 열 manufacturer와 model을 추가하고, name 열은 삭제하시오.

```python
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:]))
mpg.drop(columns=['name'], inplace=True)
mpg.head()
# column(series)에 apply를 적용하여 x가 element가 되도록
```

c. 엔진의 실린더(cylinders) 갯수별 연비(mpg)의 평균을 구하시오.

```python
mpg.groupby('cylinders')['mpg'].mean()
```

d. 생산지(origin)별 배기량 대비 마력(hp_per_cc)의 평균을 구하시오.

```python
mpg.groupby('origin')[['mpg']].mean()
# [['mpg']] '[]'를 두 개로 감싸면 Data Frame
```
